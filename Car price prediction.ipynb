```python
predict = "price"
data = data[["symboling", "wheelbase", "carlength", "carwidth", "carheight", "curbweight", "enginesize", "boreratio", "stroke", 
             "compressionratio", "horsepower", "peakrpm", "citympg", "highwaympg", "price"]]
x = x = np.array(data.drop(labels=[predict], axis=1))
y = np.array(data[predict])

from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2)

from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(xtrain, ytrain)
predictions = model.predict(xtest)

from sklearn.metrics import mean_absolute_error
model.score(xtest, predictions)
```

```python
import matplotlib.pyplot as plt
predictions = model.predict(xtest)
plt.figure(figsize=(10, 6))
plt.scatter(ytest, predictions, alpha=0.5)
plt.plot([min(ytest), max(ytest)], [min(ytest), max(ytest)], 'r--')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual vs. Predicted Car Prices')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# Create an instance of the StandardScaler
scaler = StandardScaler()

# Fit the scaler on the training data
scaler.fit(xtrain)

# Transform the training data using the scaler
xtrain_scaled = scaler.transform(xtrain)

# Transform the test data (or custom data) using the scaler
xtest_scaled = scaler.transform(xtest)

# Custom data values
custom_data = [[3, 88.6, 168.8, 64.1, 48.8, 2548, 130, 3.47, 2.68, 9, 111, 9000, 21, 27]]

# Scale the custom data using the scaler
custom_data_scaled = scaler.transform(custom_data)

# Make predictions on the custom data
custom_predictions = model.predict(custom_data_scaled)

# Print the predicted price
print("Predicted Price:", custom_predictions[0])
```
